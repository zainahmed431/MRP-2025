Outlier Detection


```python
#1st Year
outlier_counts = {}

for col in df1.columns[:-1]:  # Exclude 'class'
    Q1 = df1[col].quantile(0.25)
    Q3 = df1[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df1[(df1[col] < lower) | (df1[col] > upper)]
    outlier_counts[col] = len(outliers)

# Sort and view top features with most outliers
import pandas as pd
outlier_summary = pd.Series(outlier_counts).sort_values(ascending=False)
print(outlier_summary)
```

```python
feature='Attr10'

plt.figure(figsize=(6, 4))
sns.boxplot(y=df1[feature])
plt.title(f"Boxplot of {feature}")
plt.axhline(lower, color='red', linestyle='--', label='Lower Bound')
plt.axhline(upper, color='red', linestyle='--', label='Upper Bound')
plt.legend()
plt.tight_layout()
plt.show()
```

PCA Visualization

```python
from sklearn.impute import SimpleImputer
```

```python
from sklearn.impute import KNNImputer
```

```python
# Step 1: Separate features from target
X1 = df1.drop(columns=['class'])
y1 = df1['class']

# Step 2: Apply KNN Imputation
imputer = KNNImputer(n_neighbors=5)  # We can adjust the number of neighbors
X1_imputed = imputer.fit_transform(X1)

# Step 3: Standardize the data
scaler = StandardScaler()
X1_scaled = scaler.fit_transform(X1_imputed)

# Step 4: Apply PCA
pca = PCA(n_components=2)
X1_pca = pca.fit_transform(X1_scaled)

# Step 5: Plot the result
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X1_pca[:, 0], y=X1_pca[:, 1], hue=y1, palette='Set1', alpha=0.6)
plt.title("2D PCA Projection After KNN Imputation for 1st Year")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")

plt.show()
```

Feature Importance (Random Forest)


```python
from sklearn.ensemble import RandomForestClassifier
```
